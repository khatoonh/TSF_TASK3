# Data Visualization

```python
cm=df.corr()
cm
```

```python
sns.heatmap(cm,annot=True)
plt.show()
```

## Thus we can conclude that Sales and Profits are highly corelated  and Discounts and Profits are negatively corelated.

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.barplot(x="Ship Mode",y="Sales",data=df)
plt.title("Ship Mode v/s Sales")
plt.subplot(1,2,2)
sns.barplot(x="Ship Mode",y="Profit",data=df)
plt.title("Ship Mode v/s Profit")
plt.show()
```

```python
sns.boxplot(x='Segment',y='Quantity',data=df,hue='Ship Mode')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.barplot(x="Region",y="Sales",data=df)
plt.title("Region v/s Sales")
plt.subplot(1,2,2)
sns.barplot(x="Region",y="Profit",data=df)
plt.title("Region v/s Profit")
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
sns.barplot(x="Category",y="Sales",data=df)
plt.title("Category v/s Sales")
plt.subplot(1,2,2)
sns.barplot(x="Category",y="Profit",data=df)
plt.title("Category v/s Profit")
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,5))
df.groupby('Sub-Category')['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[0]).set_title("quantity & Profit Relation based on Sub-Category")
df.groupby('Sub-Category')['Profit','Discount'].agg(sum).plot(kind='bar',ax=axes[1]).set_title("quantity & Profit Relation based on Sub-Category")  
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.barplot(x="Segment",y="Sales",data=df)
plt.title("Segment v/s Sales")
plt.subplot(1,2,2)
sns.barplot(x="Segment",y="Profit",data=df)
plt.title("Segment v/s Profit")
plt.show()
```

```python
sns.lineplot(x='Discount',y='Sales',data=df,hue="Category",style='Category',markers=True)
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',data=df,hue='Category',style='Category',markers=True)
plt.show()
```

```python
sns.distplot(df['Quantity'],color="green")
plt.show()
```

# Some important findings

#### 1. The features Sales and Profits are highly related thus increase in sales would increase the profit.
#### 2.The maximum quantity of product in demand are in range 2-4.
#### 3.The mode of shipping doesn't affect much to the sales and profits.
#### 4.Technology related products provide high sales followed by high profits.
#### 5.Discounts increase the sales of technology related product and decreases the profit. Sales are in peak  when discount is 50%.
#### 6.Profit from central region is less.
